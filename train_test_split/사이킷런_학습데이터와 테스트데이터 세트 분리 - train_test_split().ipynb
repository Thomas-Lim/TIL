# 학습데이터와 테스트데이터 세트 분리 - train_test_split()

```python
# train_test_split()은 첫 번째 파라미터로 피처 데이터 세트, 
#                      두 번째 파라미터로 레이블 데이터 세트를 입력받는다         
#                      그리고 기타 옵션과 함께 사용된다.
```

```python
# 학습세트를 70%, 테스트데이터세트를 30%로 분리
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split # 이 부분이 그 역할 하는 것

dt_clf = DecisionTreeClassifier()
iris_data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.3, random_state=121)

# test_size는 전체 데이터에서 Test데이터 세트 크기를 얼마로 샘플링할 것인가를 결정. 디폴트는 0.25, 즉 25%. 여기서는 30%로 설정
# random_state = 호출할 때 마다 동일한 학습/테스트용 데이터 세트를 생성하기 위해 주어지는 난수 값. 
#         이를 지정하지 않으면 수행할 때마다 다른 학습/테스트 용 데이터를 생성한다. 값 지정시 수행할 때마다 동일한 데이터 세트로 분리한다.
```

```python
# 이를 기반으로 결정트리분류를 학습하고, 이 모델을 바탕으로 테스트를 해 예측 정확도 확인
#dt_clf.fit(X_train, y_train)
#pred = dt_clf.predict(X_test, y_test)     <- 이게 안되는 이유는?????
#print('예측 정확도: {0:.4f}'.format(accuracy_score(pred)))  <- 이게 안되는 이유는?????
# fit은 피처 분류(라벨) 등을 의미한다.

dt_clf.fit(X_train, y_train)
pred = dt_clf.predict(X_test)
print('예측 정확도: {0:.4f}'.format(accuracy_score(pred, y_test)))
```

```python
# 붓꽃 데이터는 150개의 데이터로 데이터 양이 크지 않아 전체의 30% 정도인 테스트 데이터는 45개 정도밖에 되지 않으므로, 
# 이를 통해 알고리즘의 예측 성능을 판단하기엔 무리가 있다.
```
